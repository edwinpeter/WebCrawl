```python
playing = pd.read_csv('cricket.csv', encoding='utf-8')
playing
```

```python
playing['Batting and fielding averages'] = ''
playing['Bowling averages'] = ''
playing
```

```python
driver = webdriver.Firefox(executable_path = '/usr/local/bin/geckodriver')
driver.implicitly_wait(10)

for i in range(len(playing)):
    playerid = playing['id'][i]
    print(playerid)
    url = 'http://www.espncricinfo.com/india/content/player/' + str(playerid) + '.html'
    driver.get(url)
    a = driver.find_elements_by_class_name("engineTable")
    bat = a[0].text.split('\n')
    bowl = a[1].text.split('\n')
    playing['Batting and fielding averages'].values[i] = bat
    playing['Bowling averages'].values[i] = bowl
```

```python
playing.to_csv('crawled.csv')
```
